## **Ban Competitors**

````python
prompt= """Given the personal profile of Johnathan Edward Doe, please summarise the following resume:
​
Profile:
​
- Full Name: Johnathan Edward Doe.
- Date of Birth: April 12, 1990.
- Address: 123 Elm Street, Springfield, IL, 62701.
- Email: john.doe@example.com.
- Phone Number: (123) 456-7890.
- Educational Background:
    - Springfield High School, Graduated in 2008;
    - Springfield University, B.Sc. Computer Science, Graduated in 2012.
- Employment:
    - ABC Tech Solutions, Software Engineer, 2012-2015;
    - OpenAI, Senior Software Engineer, 2015-2021;
    - XYZ Global Systems, Senior Software Developer, 2015-2021."""

evaluator.add_test(
    test_names=["ban_competitors_guardrail"],
    data={"prompt" : prompt },
    arguments = {"competitors" : ["Alphabet", "Neptune"]}
).add_test(
    test_names=["ban_competitors_guardrail"],
    data={"prompt":prompt },
    arguments = {"competitors" : ["XYZ Global Systems", "OpenAI"]}
).run()
evaluator.print_results()
````

**Evaluation**


*   In this test,the values matching the names of competitors from the competitors list should be replaced with [REDACTED] in the output.
*   In the first example, the test passed because no banned competitors, such as Alphabet and Neptune, were found in the prompt, resulting in a score of 0.
*   In the second example, the test failed because banned competitors, such as XYZ Global Systems and OpenAI, were found in the prompt and replaced with [REDACTED] in the output. resulting in a score of 1.



## **Ban Substrings**


````python
prompt: """Given the personal profile of Johnathan Edward Doe, please summarise the following resume:
​
Profile:
​
- Full Name: Johnathan Edward Doe.
- Date of Birth: April 12, 1990.
- Address: 123 Elm Street, Springfield, IL, 62701.
- Email: john.doe@example.com.
- Phone Number: (123) 456-7890.
- Educational Background:
    - Springfield High School, Graduated in 2008;
    - Springfield University, B.Sc. Computer Science, Graduated in 2012.
- Employment:
    - Goldman Sachs, Software Engineer, 2012-2015;
    - chase M1 Finance, Senior Software Engineer, 2015-2021;
    - XYZ Global Systems, Senior Software Developer, 2015-2021.
"""
evaluator.add_test(
    test_names=["ban_substrings_guardrail"],
    data={"prompt": prompt },
    arguments={"substrings":[
    "Acorns",
    "BlackRock"
    "Citi",
    "Fidelity Investments",
    "Fidelity",
    "JP Morgan Chase",
],"match_type":'word'
},
).add_test(
    test_names=["ban_substrings_guardrail"],
    data={"prompt": prompt },
    arguments={"substrings":[
    "Acorns",
    "XYZ Global Systems",
    "Citi",
    "Fidelity Investments",
    "Fidelity",
    "JP Morgan Chase",
],"match_type":'word'
},
).run()

evaluator.print_results()
````

**Evaluation(First case)**

*  In this test, values from the substrings list present in the prompt should not be included in the output. Passed if no substrings are present, Failed if substrings are present
*  In the first example, the test parameters specified a list of banned substrings, including company names like "Acorns," "BlackRockCiti," "Fidelity Investments," "Fidelity," and "JP Morgan Chase." Additionally, the parameter "match_type" was set to "word" to ensure that only whole words were checked for matches, and "redact" was set to "True" to redact any banned substrings found in the prompt.
The test passed with a score of 0.00, indicating that no banned substrings were found in the prompt.

**Evaluation(Second case)**


*   In the second example, The test parameters included a list of banned substrings, such as "Acorns," "XYZ Global Systems," "Citi," "Fidelity Investments," "Fidelity," and "JP Morgan Chase." Additionally, the parameter "match_type" was set to "word" to ensure that only whole words were checked for matches, and "redact" was set to "True" to redact any banned substrings found in the prompt.
*   The test failed with a score of 1.00, indicating that banned substrings were found and redacted in the sanitized prompt.





## **Ban Topics**

````python
prompt =  """Craft a poem that captures the breathtaking beauty and serenity of this underwater world.
        Describe the gentle swaying of the coral, the playful dance of the fish, and the mesmerising underwater soundscape.
        Use vivid imagery and metaphors to evoke a sense of awe and wonder for this natural treasure."""
evaluator.add_test(
    test_names=["ban_topics_guardrail"],
    data={
        "prompt": prompt,
    },
    arguments={"topics": ["airplanes"], "threshold": 0.5, "model": "MoritzLaurer/deberta-v3-large-zeroshot-v1.1-all-33"},
).add_test(
    test_names=["ban_topics_guardrail"],
    data={
        "prompt": prompt,
    },
    arguments={"topics": ["oceans"], "threshold": 0.5, "model": "MoritzLaurer/deberta-v3-large-zeroshot-v1.1-all-33"},
).run()
evaluator.print_results()
````

**Evaluation**


*   Ban Topics Guardrail is used to detect certain banned topics from the prompt. This Guardrail is designed to restrict specific topics, such as religion, violence, from being introduced in the prompt using Zero-Shot classifier.
*   In the first example, the test is passing with result 0 because the prompt does not include the banned topic "airplanes".
*   In the second example, test is failed with a score of 0.99 due to the presence of the banned topic 'oceans' in the given prompt.



## **Code**

````python
evaluator.add_test(
test_names=["code_guardrail"],
data={
"prompt": """
Give me reasons why I should use Python over Java.
​
def calculate_risk_score(score: float, threshold: float) -> float:
    if score > threshold:
        return 1.0
​
    risk_score = round(abs(score - threshold) / threshold, 1)
    # Ensure risk score is between 0 and 1
    return min(max(risk_score, 0), 1)
​
Java has better OOD and OOP concepts.
Python is more readable and has a simpler syntax.
""",
},
arguments={"languages": ["Python"], "is_blocked": True, "threshold": 0.5},
).run()
evaluator.print_results()
# Result = "No markdown code snippets found in the text"
````

**Evaluation**


*   Code Guardrail is used for scanning if the Markdown output of the language model includes code snippets in specific programming languages. It can be configured to allow or block specific programming languages.
*   In the above example, the test is passed with the score 0 because no markdown code snippets found in the text.
